```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## manual matrices classwise K
## RGB, dense
ks = [1, 2, 3]
c_1 = [.640339745035369,0.6442625037341537, 0.6305543346519343]
c_2 = [0.3069412249872155,0.2916062102116699, 0.2931648273686814]
c_3 = [0.5200599390757321,0.5298656444785972, 0.5359303405653624]
c_4 = [0.2951659878741631,0.31081015731917727, 0.30012532298555994]
c_5 = [0.5554777168226458,0.5608815182939524, 0.564672497304779]

means = [((c_1[i]+c_2[i]+c_3[i]+c_4[i]+c_5[i])/5) for i in range(len(c_1))]
print (means)
fig = plt.figure(figsize=(10, 5))
plt.plot(ks,c_1,label='airplanes')
plt.plot(ks,c_2, label='birds')
plt.plot(ks,c_3, label='cars')
plt.plot(ks,c_4,label='horses')
plt.plot(ks,c_5,label='ships')
plt.plot(ks,means,label='mean')
plt.plot()
plt.xticks([1, 2, 3], ['400', '1000', '4000'])
plt.legend()
plt.show()
fig.savefig('k_variation.png')
```
